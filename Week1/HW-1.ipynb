````python
X = df_temp.drop_duplicates().reset_index(drop=True).values
````

````python
X
````

````python
XT = np.transpose(X)
XT
````

````python
XTX = XT.dot(X)
````

````python
XTX
````

````python
XTX.shape
````

````python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
````

````python
print(XTX_inv.shape, XT.shape)
````

````python
y = [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800]

temp = XTX_inv.dot(XT)

w = temp.dot(y)
````
